```python
import pandas as pd
from sklearn import linear_model
import matplotlib as mp
import matplotlib.pyplot as plt

df = pd.read_csv("insurance.csv")

```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['age', 'sex', 'bmi', 'children', 'smoker']], df.charges)
```
